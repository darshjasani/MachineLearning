```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df.dtypes
```

```python
df.corr() #correlation 
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x="petal_length",y="petal_width",data=df,kind="resid")
```

```python
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='kde',hue='species')
```

```python
sns.pairplot(df)
```

```python
sns.displot(df['sepal_length'],kde = True)
```

```python
sns.countplot(y='sepal_width',data=df)
```

```python
sns.barplot(y='sepal_length',x='petal_length',data=df,hue='species')
```

```python
sns.barplot(y='sepal_length',x='petal_length',data=df,hue='species',ci=None)
```

```python
sns.boxplot(x='petal_length',y='petal_width',data=df)
```

```python
sns.violinplot(data=df,x='petal_length',y='petal_width')
```
