```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.corr()
```

## Univariate Analysis

Predict Output using one feature

```python
df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Sepal Length')
plt.show()
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros(len(df_setosa['sepal_length'])),'o')
```

```python
plt.plot(df['sepal_length'],np.zeros_like(df['sepal_length']),'o')
```

```python
sns.scatterplot(data=df,x='petal_length',y=np.zeros_like(df['petal_length']),hue='species')
```

## Bivariate Analysis

```python
sns.scatterplot(data=df,x='petal_length',y='petal_width',hue='species')
```

```python
sns.FacetGrid(data=df,hue='species',size=5).map(plt.scatter,'petal_length','petal_width').add_legend()
plt.show()
```

```python
plt.scatter(df_setosa['petal_length'],df_setosa['petal_width'],c='blue',marker='o')
plt.scatter(df_virginica['petal_length'],df_virginica['petal_width'],c='green',marker='o')
plt.scatter(df_versicolor['petal_length'],df_versicolor['petal_width'],c='orange',marker='o')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend(['setosa','virginica','versicolor'])
plt.show()
```
